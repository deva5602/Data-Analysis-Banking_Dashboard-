```python
#generating descriptive statistics for the data frame 
df.describe()
```

```python
# estimated income 
df['Estimated Income'].value_counts()
```

```python
df['Estimated Income'].min()
# now i want to convert this numerical data into categorical one eg= 0 to 10000 , 10000 to 20000 etc.

bins = [0 , 100000 , 300000 , float('inf')]
labels = ['low', 'medium' , 'high']

df['Income Band'] = pd.cut(df['Estimated Income'] , bins= bins , labels=labels  , right = False )
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
#examine the distribution of unique cataegories in categorical columns
# also called as univariate analysis i,e one column at a time 
categorical_col = df[['BRId' , 'GenderId' ,'IAId' , 'Amount of Credit Cards' , 'Nationality' , 'Occupation' , 'Fee Structure' , 'Loyalty Classification' , 'Properties Owned' , 'Risk Weighting' , 'Income Band']].columns

for col in categorical_col:
    print(f"value counts for '{col}':")
    display(df[col].value_counts())
```

```python
for i , predictor in enumerate(df[['BRId' , 'GenderId' ,'IAId' , 'Amount of Credit Cards' , 'Nationality' , 'Occupation' , 'Fee Structure' , 'Loyalty Classification' , 'Properties Owned' , 'Risk Weighting' , 'Income Band']]):
    plt.figure(i)
    sns.countplot(data = df , x= predictor , hue= 'GenderId')

# if a remobe hue it will become univariate analysis with hue it is bivariate
```

```python
#analysis with respect to nationality 
for i , predictor in enumerate(df[['BRId' , 'GenderId' ,'IAId' , 'Amount of Credit Cards' , 'Nationality' , 'Occupation' , 'Fee Structure' , 'Loyalty Classification' , 'Properties Owned' , 'Risk Weighting' , 'Income Band']]):
    plt.figure(i)
    sns.countplot(data = df , x= predictor , hue= 'Nationality')
```

```python
#histplot of value counts for diffrent occupation 

for col in categorical_col:
    if col == 'Occupation':
        continue 
    plt.figure(figsize=(8,5))
    #plt.subplot(5,3 , i+1 )
    sns.histplot(df[col])
    plt.title ('histogram of occupation count')
    plt.xlabel(col)
    plt.ylabel('count')
    plt.show()
```

#numeric analysis 

```python
numeric_col = ['Estimated Income' ,	'Superannuation Savings' ,	'Credit Card Balance' ,	'Bank Loans' ,	'Bank Deposits' ,	'Checking Accounts' ,	'Saving Accounts' ,	'Foreign Currency Account' ,	'Business Lending' ]

#univariate Analysis 
plt.figure(figsize=(20,10))
for i , col in enumerate(numeric_col):
    plt.subplot(4,3 , i+1)
    sns.histplot(df[col] ,kde=True )
    plt.title(col)
plt.show()
```

#Heat maps 

```python
numeric_col = ['Estimated Income' ,	'Superannuation Savings' ,	'Credit Card Balance' ,	'Bank Loans' ,	'Bank Deposits' ,	'Checking Accounts' ,	'Saving Accounts' ,	'Foreign Currency Account' ,	'Business Lending' ]

correlation_matrix = df[numeric_col].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix , annot=True , cmap='crest' , fmt='.2f')
plt.title('correlation Matrix')
plt.show()
```
